```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
gau = lambda mu, s2: np.exp(-(mu**2)/(2*s2)) / np.sqrt(2*np.pi*s2)

N = 100
x = np.linspace(0, 1, N)
y = gau(x - 0.5, 1)
    
plt.figure()
plt.plot(x, y)
plt.show()
```

```python
polar = [(y[i], 2*np.pi*i/N) for i in range(N)]
cartesian = [(r*np.cos(phi), r*np.sin(phi)) for r, phi in polar]

cartesian = np.array(cartesian).T
cMean = np.mean(cartesian, axis=1)

plt.figure()
plt.plot(*cartesian)
plt.arrow(0, 0, *cMean)

plt.show()
```

```python
r = np.linalg.norm(cMean)
phi = np.arctan2(cMean[1], cMean[0])
if phi < 0:
    phi += 2 *np.pi
```

```python
plt.figure()
plt.plot(x, y)
plt.axvline(x=phi / 2 / np.pi)
plt.show()
```
